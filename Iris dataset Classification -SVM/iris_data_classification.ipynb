```python
from sklearn import datasets # to lead the default datasets 
from sklearn import metrics # parameter evaluation 
from sklearn import svm # default classfier
```

```python
# load the data 
data = datasets.load_iris()
print(type(data))

X = data.data # accessing the data from the utils bunch 
Y = data.target # respective outputs of the data

print(X.shape)
print(Y.shape)
print(data.feature_names)
print(data.target_names)
print(Y)
print(X)
```

```python
# split of the dataset into train data and test data
from sklearn import model_selection # will be used to split the data into train:test ratio 
# 70:30 , 80:20, 60:40 
# 0.3 - 30% testing ratio
[Xtrain,Xtest,ytrain,ytest]=model_selection.train_test_split(X,Y,test_size=0.5,random_state=10)
print(Xtrain.shape,ytrain.shape)
print(Xtest.shape,ytest.shape)
```

```python
# create the classification model 
svmModel = svm.SVC() # learning model 

# train the model using the trianing data 
svmModel = svmModel.fit(Xtrain,ytrain) # passign the train data and its respective output (target)


# test the trained model using only test data 
op = svmModel.predict(Xtest) #only passing the test data for the output (target prediction )



```

```python
# performance evaluations
acc = metrics.accuracy_score(op,ytest)
print("Accuracy of the model is : ",acc)

conf_matrix = metrics.confusion_matrix(op,ytest)
print("Confusion Matrix: ")
print(conf_matrix)

report = metrics.classification_report(op,ytest)
print("complete report of the iris dataset")
print(report)
```
