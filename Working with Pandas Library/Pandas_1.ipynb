```python
result = np.zeros((5,3))
year = [2011,2012,2013,2014,2015]
plat = ['PC','PlayStation 3','Xbox 360']

for i in range(5):
    for j in range(3):
        ind = (data['release_year']==year[i]) &(data['platform']==plat[j])
        new_data = data[ind]
        f = new_data['platform'].value_counts()
        result[i,j]=f
print(result)

print(type(result))

resultDF = pd.DataFrame(result,columns = plat,index=year)
print(resultDF)
print(type(resultDF))

# plotting using .plot function of pandas 
resultDF.plot(kind='bar')
```

```python
result = np.zeros((5,3))
phrase = ['Great','Okay','Mediocre','Amazing','Masterpiece']
plat = ['PC','PlayStation 3','Xbox 360']

for i in range(result.shape[0]):
    for j in range(result.shape[1]):
        ind = (data['score_phrase']==phrase[i]) &(data['platform']==plat[j])
        new_data = data[ind]
        f = new_data['platform'].value_counts()
        result[i,j]=f
print(result)
print(type(result))

resultDF = pd.DataFrame(result,columns = plat,index=phrase)
print(resultDF)
print(type(resultDF))

# plotting using .plot function of pandas 
resultDF.plot(kind='bar')
```

```python
# from user ask the platform and plot the scorephrase wise comparative bar chart 
pl = input("Enter the Platform Name: ")
print("Platform Entered is: ",pl)

ind = data['platform']==pl
newData=data[ind]

col = data['score_phrase'].unique()
print(col)


res = np.zeros((1,len(col)))
print(res.shape)
for i in range(len(col)):
    ind1 = newData['score_phrase']==col[i]
    val = newData[ind1]
    res[0,i]=val.shape[0]
print(res)

resDF = pd.DataFrame(res,index=[pl],columns = col)
resDF = resDF.T
resDF.plot(kind='bar')
```
