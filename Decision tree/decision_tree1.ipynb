```python
from sklearn import tree
from sklearn import metrics,datasets
from sklearn.model_selection import train_test_split

# iris dataset 
data = datasets.load_iris()

X = data.data
y =data.target

print(data.target_names)
print(data.feature_names)
print(data.data[1,:])
```

```python
Xtrain,Xtest,ytrain,ytest = train_test_split(X,y,test_size=0.3,random_state=2)

# create the Dtree 

DtModel = tree.DecisionTreeClassifier(criterion='gini',random_state=2)
DtModel = DtModel.fit(Xtrain,ytrain)

opTrain=DtModel.predict(Xtrain) # to calculated the training acccuracy 
optest= DtModel.predict(Xtest)



# performance evaluation 
Trainacc = metrics.accuracy_score(opTrain,ytrain) 
Testacc = metrics.accuracy_score(optest,ytest)

print("Training Accuracy: ",Trainacc)
print("Testing Accuracy: ",Testacc)
```

```python
import matplotlib.pyplot as plt 

plt.figure(10,figsize=(12,12))
tree.plot_tree(DtModel,filled=True,rounded=True)
```
