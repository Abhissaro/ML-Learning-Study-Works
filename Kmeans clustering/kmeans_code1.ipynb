```python
from sklearn import cluster
import numpy as np 
import matplotlib.pyplot as plt

samples= 500
data = np.random.randint(0,100,size=(500,2))
print(data.shape)
plt.figure(1)
# markers o-cicle,s-square,d-diamond,*-star
plt.plot(data[:,0],data[:,1],'ko')
plt.title('raw data')
```

```python
# kmeans models for 2 cluster 
kmModel = cluster.KMeans(n_clusters=2,random_state=10)
kmModel = kmModel.fit(data)
print(kmModel.cluster_centers_)
print(kmModel.labels_)
```

```python
label = kmModel.labels_
centroids = kmModel.cluster_centers_

plt.figure(1,figsize=(9,9))
plt.subplot(2,1,1)
# markers o-cicle,s-square,d-diamond,*-star
plt.plot(data[:,0],data[:,1],'ko')
plt.title('raw data')
plt.subplot(2,1,2)
plt.scatter(data[:,0],data[:,1],c=label)
plt.title('Segmented Data')
plt.plot(centroids[0][0],centroids[0][1],'bd')
plt.plot(centroids[1][0],centroids[1][1],'bd')
```

```python
# kmeans models for 4 cluster 
kmModel = cluster.KMeans(n_clusters=4,random_state=10)
kmModel = kmModel.fit(data)
print(kmModel.cluster_centers_)
print(kmModel.labels_)
label = kmModel.labels_
centroids = kmModel.cluster_centers_

plt.figure(1,figsize=(9,9))
plt.subplot(2,1,1)
# markers o-cicle,s-square,d-diamond,*-star
plt.plot(data[:,0],data[:,1],'ko')
plt.title('raw data')
plt.subplot(2,1,2)
plt.scatter(data[:,0],data[:,1],c=label)
plt.title('Segmented Data')
for i in range(len(centroids)):
    plt.plot(centroids[i][0],centroids[i][1],'bd')

    
kmModel.inertia_
```

```python
# elbow method
cl = []
dis = []
for i in range(1,11,1):
    cl.append(i)
    kmModel = cluster.KMeans(n_clusters=i,random_state=10)
    kmModel = kmModel.fit(data)
    dis.append(kmModel.inertia_)

plt.figure(1)
plt.plot(cl,dis,'--ko')
plt.xlabel('clusters')
plt.ylabel('sum of squared distances')
plt.grid('on')


# the best cluster would be 4 as clearly seen from the elbow graph 
    

```

```python
import pandas as pd


data = pd.read_csv('./Salary_Data.csv')
print(data.head())
print(data.columns)
X = data.values
```

```python
# elbow method
cl = []
dis = []
for i in range(1,11,1):
    cl.append(i)
    kmModel = cluster.KMeans(n_clusters=i,random_state=10)
    kmModel = kmModel.fit(X)
    dis.append(kmModel.inertia_)

plt.figure(1)
plt.plot(cl,dis,'--ko')
plt.xlabel('clusters')
plt.ylabel('sum of squared distances')
plt.grid('on')

# final number of clusters is 3
# kmeans models for 3 cluster 
kmModel = cluster.KMeans(n_clusters=3,random_state=10)
kmModel = kmModel.fit(X)
print(kmModel.cluster_centers_)
print(kmModel.labels_)
label = kmModel.labels_
centroids = kmModel.cluster_centers_

plt.figure(2,figsize=(9,9))
plt.subplot(2,1,1)
# markers o-cicle,s-square,d-diamond,*-star
plt.plot(X[:,0],X[:,1],'ko')
plt.title('raw data')
plt.subplot(2,1,2)
plt.scatter(X[:,0],X[:,1],c=label)
plt.title('Segmented Data')
for i in range(len(centroids)):
    plt.plot(centroids[i][0],centroids[i][1],'bd')
```
