```python
# unsupervised learning algorithm 
import numpy as np
import matplotlib.pyplot as plt 

sample= 500 
data = np.random.randint(0,100,size=(sample,2))
#print(data)

plt.figure(1)
plt.plot(data[:,0],data[:,1],'ko')
plt.title('raw data')

```

```python
# kmeans 
from sklearn import cluster 

# create the kmeans model 
kmmodel = cluster.KMeans(n_clusters=2,random_state=2)
kmmodel = kmmodel.fit(data)
#print(kmmodel.labels_)
plt.figure(1,figsize=(10,10))
plt.subplot(2,1,1)
plt.plot(data[:,0],data[:,1],'ko')
plt.title('raw data')

plt.subplot(2,1,2)
plt.scatter(data[:,0],data[:,1],c=kmmodel.labels_)
plt.title('segmented data')

kmmodel.inertia_
```

```python
# elbow method on this random data 

cl = []
dist=[]

for i in range(1,11):
    # create the kmeans model 
    kmmodel = cluster.KMeans(n_clusters=i,random_state=2)
    kmmodel = kmmodel.fit(data)
    cl.append(i)
    dist.append(kmmodel.inertia_)

plt.figure(1,figsize=(10,8))
plt.plot(cl,dist,'--ko')
plt.title('elbow method')
plt.grid('on')
    
    
```

```python
# form the elbow method, the best number of clusters identified are 4

# create the kmeans model 
kmmodel = cluster.KMeans(n_clusters=4,random_state=2)
kmmodel = kmmodel.fit(data)
#print(kmmodel.labels_)
plt.figure(1,figsize=(10,10))
plt.subplot(2,1,1)
plt.plot(data[:,0],data[:,1],'ko')
plt.title('raw data')

plt.subplot(2,1,2)
plt.scatter(data[:,0],data[:,1],c=kmmodel.labels_)
plt.title('segmented data')

```

```python
npath = "./Mall_Customers.csv"
ndata2 = pd.read_csv(npath)
newdata = ndata2.drop(["Genre"],axis=1)
newdata.head()
```

```python
kmmodel = cluster.KMeans(n_clusters=5,random_state=2)
newdata.drop(['Age','CustomerID'],axis=1,inplace=True)
kmmodel = kmmodel.fit(newdata)
#print(kmmodel.labels_)
plt.figure(2,figsize=(10,10))
plt.subplot(2,1,1)
plt.plot(newdata.iloc[:,0],newdata.iloc[:,1],'ko')
plt.title('salary data')

plt.subplot(2,1,2)
plt.scatter(newdata.iloc[:,0],newdata.iloc[:,1],c=kmmodel.labels_)
plt.title('segmented data')
```

```python
cl = []
dist=[]

for i in range(1,21):
    # create the kmeans model 
    kmmodel = cluster.KMeans(n_clusters=i,random_state=2)
    kmmodel = kmmodel.fit(newdata)
    cl.append(i)
    dist.append(kmmodel.inertia_)

plt.figure(1,figsize=(10,8))
plt.plot(cl,dist,'--ko')
plt.title('elbow method')
plt.grid('on')
```
